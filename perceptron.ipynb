```python
import numpy as np

# Training data: AND gate
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 0, 0, 1])   # AND labels
```

```python
# Perceptron parameters
weights = np.zeros(2)
bias    = 0
lr      = 0.1           # learning rate
epochs  = 10
```

```python
weights.shape
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        z    = np.dot(weights, xi) + bias
        pred = step(z)
        err  = yi - pred            # error
        weights += lr * err * xi    # weight update
        bias    += lr * err
        errors  += int(err != 0)
    print(f"Epoch {epoch+1}: errors={errors}, w={weights}, b={bias:.1f}")
```

```python
# Test
print("\n--- Predictions ---")
for xi, yi in zip(X, y):
    pred = step(np.dot(weights, xi) + bias)
    print(f"Input {xi} → pred={pred}, actual={yi}")
```

MLP FOR XOR


```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])   # XOR labels
```

```python
# Architecture: 2 → 2 → 1
np.random.seed(42)
W1 = np.random.randn(2, 2) * 0.5
b1 = np.zeros((1, 2))
W2 = np.random.randn(2, 1) * 0.5
b2 = np.zeros((1, 1))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_deriv(a):
    return a * (1 - a)
```

```python
lr = 0.5
```

```python
for epoch in range(10000): # Increased epochs for better XOR convergence
    # 1. Forward pass (Use @ for matrix multiplication)
    Z1 = X @ W1 + b1    
    A1 = sigmoid(Z1)
    Z2 = A1 @ W2 + b2
    A2 = sigmoid(Z2)

    # 2. Compute Loss
    loss = np.mean((y - A2) ** 2)

    # 3. Backpropagation (Must be INSIDE the loop)
    dA2 = (A2 - y) # Simplified derivative for MSE + Sigmoid logic
    dZ2 = dA2 * sigmoid_deriv(A2)
    dW2 = A1.T @ dZ2
    db2 = np.sum(dZ2, axis=0, keepdims=True)

    dA1 = dZ2 @ W2.T
    dZ1 = dA1 * sigmoid_deriv(A1)
    dW1 = X.T @ dZ1
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # 4. Gradient descent
    W2 -= lr * dW2
    b2 -= lr * db2
    W1 -= lr * dW1
    b1 -= lr * db1

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}: loss={loss:.4f}")
```
